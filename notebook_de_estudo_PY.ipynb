```python

##################################
##  criaГ§ГЈo da classe e objetos ##
##################################

class Pessoa:
  def __init__(self, n, id, tel, uf):
      self.nome = n
      self.cpf = id
      self.contato = tel
      self.cidade = uf

class Funcionario(Pessoa):
  def __init__ (self, n, id, tel, uf, salario, cargo):
      super().__init__(n, id, tel, uf)
      self.salario = salario
      self.cargo = cargo

  def exibir_cracha(self):
      print("\n")
      print(f"-- Nome: {self.nome} ")
      print(f"-- Cargo: {self.cargo} ")
      print(f"-- SalГЎrio: R$ {self.salario} ")
      print("--"*20)


######################################
##  Cadastranto objetos nas classes ##
######################################

f1 = Funcionario("Carlos Roberto", "444-555","9888-7777", "MT", 9999.00,"Analista")

f2 = Funcionario("Solrac Coelho", "123-456","9876-5432", "RO", 5000.00, "Analista RO")

f3 = Funcionario("Solrac Coelho", "789-456", "7418-8529", "AC",'' ,"Analista RO")


####################################
##  Exibindo os itens cadastradis ##
####################################

f1.exibir_cracha()
f2.exibir_cracha()
f3.exibir_cracha()
```
